# HW Question 5

```python
query="""
SELECT count(lpep_pickup_datetime)
FROM green_taxi_tripdata
WHERE EXTRACT(Month FROM lpep_pickup_datetime) = 9 
AND EXTRACT(YEAR FROM lpep_pickup_datetime) = 2019 
AND EXTRACT(DAY FROM lpep_pickup_datetime) = 18
"""

pd.read_sql(query, con=engine)
```

```python
query="""
SELECT l."Borough" 
FROM green_taxi_tripdata g
INNER JOIN lookup_data l ON g."PULocationID" = l."LocationID"
GROUP BY l."Borough"
ORDER BY sum(total_amount) DESC 
LIMIT 3
;
"""

pd.read_sql(query, con=engine)
```

# HW Question 6

```python
query="""
SELECT "Zone"
FROM lookup_data
WHERE "LocationID" = (SELECT g."DOLocationID"
                        FROM lookup_data l
                        JOIN green_taxi_tripdata g 
                        ON g."PULocationID" = l."LocationID"
                        WHERE l."Zone" = 'Astoria' 
                        AND EXTRACT(YEAR FROM g.lpep_pickup_datetime) = 2019 
                        AND EXTRACT(Month FROM g.lpep_pickup_datetime) = 9 
                        GROUP BY g."DOLocationID"
                        ORDER BY max(g."tip_amount") DESC
                        LIMIT 1)
"""

pd.read_sql(query, con=engine)
```
